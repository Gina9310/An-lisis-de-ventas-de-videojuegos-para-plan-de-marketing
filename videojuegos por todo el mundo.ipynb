<div class='alert alert-block alert-info'>
<b> DESCRIPCIÓN DEL PROYECTO </b> <a class='tocSkip'></a>
Este proyecto tiene como objetivo analizar el comportamiento de ventas de videojuegos a nivel global para identificar tendencias de mercado, plataformas más rentables y factores que influyen en el éxito de un videojuego.

El análisis se enfoca en detectar proyectos prometedores y planificar campañas publicitarias.

<div class='alert alert-block alert-info'>
<b> OBJETIVOS </b> <a class='tocSkip'></a>

* Identificar las plataformas con mayor potencial de ventas

* Analizar tendencias por año, región y género

* Evaluar la relación entre calificaciones y ventas

* Determinar diferencias significativas entre plataformas y géneros mediante pruebas estadísticas

<div class='alert alert-block alert-info'>
<b> PROCESO </b> <a class='tocSkip'></a>

1. Limpieza y preparación de datos

    * Estandarización de nombres de columnas

    * Eliminación de duplicados

    * Tratamiento de valores nulos

    * Creación de variables relevantes (ventas totales)

2. Análisis exploratorio (EDA)

    * Distribución de ventas por año

    * Análisis por plataforma y género

    * Identificación de plataformas dominantes

    * Análisis por región (NA, EU, JP)

3. Análisis estadístico

Pruebas de hipótesis (t-test) para comparar:

   * Calificaciones entre plataformas (Xbox vs PC)

   * Calificaciones entre géneros (Action vs Sports)

<div class='alert alert-block alert-info'>
<b> CONCLUSIONES


</b>  1.- las ventas totales dentro de un periodo de 36 años (1980-2016) fueron de 8,816.75 millones de dolares, de las cuales el 49% tuvieron lugar en norteamerrica, el 27% en europa y el 14% en japón y solo el 10% en algún otro lugar del mundo.

```python


# Filtrar por plataforma PS4
df_ps4 = df[df['platform'] == 'PS4']

# Eliminar nulos en reseñas y ventas
df_ps4_nonulos = df_ps4.dropna(subset=['user_score', 'critic_score', 'ventas_totales'])

# Asegurar que user_score esté en formato numérico (algunos datasets lo tienen como string)
#df_ps4_nonulos['user_score'] = pd.to_numeric(df_ps4_nonulos['user_score'], errors='coerce')
df_ps4_nonulos = df_ps4_nonulos.dropna(subset=['user_score'])


```

```python
# Estilo visual
sns.set(style="whitegrid")

# Reseñas de usuarios vs ventas
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_ps4_nonulos, x='user_score', y='ventas_totales', color='dodgerblue')
plt.title('Relación entre reseñas de usuarios y ventas (PS4)')
plt.xlabel('Puntuación de usuarios')
plt.ylabel('Ventas totales (millones)')
plt.show()

# Reseñas de críticos vs ventas
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_ps4_nonulos, x='critic_score', y='ventas_totales', color='darkorange')
plt.title('Relación entre reseñas de críticos y ventas (PS4)')
plt.xlabel('Puntuación de críticos')
plt.ylabel('Ventas totales (millones)')
plt.show()
```

```python
cor_user = df_ps4_nonulos['user_score'].corr(df_ps4_nonulos['ventas_totales'])
cor_critic = df_ps4_nonulos['critic_score'].corr(df_ps4_nonulos['ventas_totales'])

print(f"Correlación entre reseñas de usuarios y ventas: {cor_user:.2f}")
print(f"Correlación entre reseñas de críticos y ventas: {cor_critic:.2f}")
```

```python
df["platform"].unique()
```

```python
#determinar calificacion promedio XONE
calif_xOne= df[df["platform"] == "XOne"]["user_score"].mean()

print(calif_xOne)
```

```python
#determinar calificacion promedio PC
calif_pc= df[df["platform"] == "PC"]["user_score"].mean()

print(calif_pc)
```

COMPARAR HIPÓTESIS:


```python
#Hipótesis nula: los promedios de la calificaciones de las dos plataformas son iguales.

#hipótesis alternativa: los promedios de las calificaciones no son iguales.
```

```python
df["user_score"] = pd.to_numeric(df["user_score"], errors="coerce")
```

```python

from scipy import stats

alpha = 0.05

# Extraer las puntuaciones individuales
xbox_scores = df[df["platform"] == "XOne"]["user_score"].dropna()
pc_scores = df[df["platform"] == "PC"]["user_score"].dropna()

# Prueba t de muestras independientes
results = stats.ttest_ind(xbox_scores, pc_scores, equal_var=False)  

print("Valor p:", results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula: hay diferencia significativa")
else:
    print("No rechazamos la hipótesis nula: no hay evidencia de diferencia")

```


<div class='alert alert-block alert-success'>
<b> Dado el resultado, se resume que el promedio de la calificaciones de los usuarios para los plataformas PC y xOne no son iguales, hay diferencia entre ellas.</b> <a class='tocSkip'></a>
</div>


<div class='alert alert-block alert-info'>
<b>  Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes. .</b> <a class='tocSkip'></a>

#Hipótesis nula: los promedios de la calificaciones de los usuarios para los géneros Acción y Deportes son iguales

#hipótesis alternativa: los promedios de la calificaciones de los usuarios para los géneros Acción y Deportes son iguales.

```python
action_scores = df[df["genre"] == "Action"]["user_score"].dropna().mean()
sports_scores = df[df["genre"] == "Sports"]["user_score"].dropna().mean()

print(action_scores)
print(sports_scores)
```

```python

alpha = 0.05

# Extraer las puntuaciones individuales
action_scores = df[df["genre"] == "Action"]["user_score"].dropna()
sports_scores = df[df["genre"] == "Sports"]["user_score"].dropna()

# Prueba t de muestras independientes
results = stats.ttest_ind(action_scores, sports_scores, equal_var=False)  

print("Valor p:", results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula: hay diferencia significativa")
else:
    print("No rechazamos la hipótesis nula: no hay evidencia de diferencia")

```

<div class='alert alert-block alert-success'>
